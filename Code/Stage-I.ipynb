### Reading data set

```python
covid_data = pd.read_csv("../../../data/covid_confirmed_usafacts.csv")
covid_data
```

### renaming the State to 'state' and County Name to 'county'

```python
covid_data.rename(columns = {'State': 'state','County Name':'county'}, inplace = True)
covid_data
```

```python
covid_data['county'] = covid_data['county'].apply(lambda cn: cn.lower().strip())
```

### Considering North Carolina 
### Now, Lets Predict wheather the cases in this region are increasing, decreasing or stable?
### Using tail() funtion to print last week data

```python
north_carolina_state_covid_cases=covid_data.loc[covid_data['state']=='NC']
last_week=north_carolina_state_covid_cases.tail(7)
last_week
```

### Using transpose() function to changes the row elements into column elements

```python
temp=north_carolina_state_covid_cases
change=temp.transpose()
change= change.sum(axis=1)
```

```python
last_week= change.tail(7)
last_week.plot(x=last_week[1], y=last_week[0], kind="line", figsize=(8,8))
```

## Enrichment Data 



### Reading governors_county.csv

```python
gv_county=pd.read_csv("../../../data/governors_county.csv")
gv_county
```


### I have used the loc fuction to exactly locate the position of the state value
### and i have used replace() function to replace the state names

```python
county_State_Shirk=(gv_county.loc[( (gv_county['state'] == 'Indiana') | gv_county['state'] == 'Delaware') | (gv_county['state'] == 'Missouri') | (gv_county['state'] == 'Montana') | (gv_county['state'] == 'South Carolina') | (gv_county['state'] == 'North Dakota') | (gv_county['state'] == 'Utah') | (gv_county['state'] == 'Vermont') | (gv_county['state'] == 'Washington') | (gv_county['state'] == 'New York') | (gv_county['state'] == 'New Hampshire')])
county_State_Shirk = county_State_Shirk.replace( ['Indiana', 'Delaware', 'Maryland', 'Missouri', 'New Hampshire','North Carolina', 'South Dakota', 'Utah', 'Vermont', 'New York','Washington'],[ 'IA','DE', 'MO', 'MI', 'NH', 'NC', 'SD', 'UT', 'VT', 'NY', 'WA'], regex=True)
county_State_Shirk
```

### Merging Enrichment data set with covid data sets

```python
county_State_Shirk['county'] = county_State_Shirk['county'].apply(lambda cn: cn.lower().strip())
```

### Using merge() funtion to merge the data sets

```python
merging_covid_enrichment_data=pd.merge(covid_data,county_State_Shirk ,on='county',how='inner')
merging_covid_enrichment_data
```
